# Matrix multiplication

Two main ways of performing multiplication in neural networks and deep learning :

1. Element-wise Multiplication
2. Matrix multiplication (dot product `(.)`)

There are two main rules that performing matrix multiplication needs to satisfy:

1. The **inner dimension** must match :
* `(3,2) @ (3,2)` won't work
* `(2,3) @ (3,2)` will work
* `(3,2) @ (2,3)` will work

2. The resulting matrix has the shape of the **outer dimension**:
* `(2,3) @ (2,2) -> (2,2)`
* `(3,2) @ (2,3) -> (3,3)`
  





```python
# Element wise multiplication
print(tensor, "*", tensor)
print(f"Equals : {tensor * tensor}")
```

## One of the most common errors in deep learning : shape error

```python
# Shapes for matrix multiplication
tensor_A = torch.tensor([[1,2],
                         [3,4],
                         [5,6]])
tensor_B = torch.tensor([[7,10],
                         [8,11],
                         [9,12]])

#torch.mm(tensor_A, tensor_B) # torch.mm is the same as torch.matmul (it's a alias for writing less code  )
torch.matmul(tensor_A, tensor_B)
```

```python
tensor_A.shape, tensor_B.shape
```

To fix our tensor shape issue, we can manipulate the shape of one of our tensor using a **transpose**

A **transpose** switches the axes or dimension of a given tensor

```python
tensor_B, tensor_B.shape
```

```python
tensor_B.T, tensor_B.T.shape
```

```python
torch.matmul(tensor_A, tensor_B.T)
```

```python
torch.matmul(tensor_A, tensor_B.T).shape
```

```python
# The matrix multiplication operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: tensor_A = {tensor_A.shape} (same shape as above), tensor_B.T = {tensor_B.T.shape}")
print(f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <- inner dimension must match")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
```

## Reshaping, stacking, squeezing and unsqueezing tensors

* Reshaping -> reshapes an input tensor to a defined shape
* View -> Return a view of an input tensor of certain shape but keep the same memory as the orginal tenosr
* Stacking -> combine multiple tensor on top of each other (Vstack) or side by side (hstack)
* Squeeze - > remove all `1` dimensions from a tensor
* Unsqueeze -> add a `1` dimensions to a target tensor
Permute -> Return a view of the input with dimensions permuted (swapped) in a certain way

```python
# Let's create a tensor
x = torch.arange(1., 10.)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1,9)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1,9)
z, z.shape
```

```python
# Changing z changes x (because a view of a tensor shares the same memory as the original input)
z[:,0] = 5
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
```

```python
# torch.sqeeze() - removes all single dimensions from a target tensor
print(f"Previous tensor : {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Remove extra dimension from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

```python
x_reshaped.shape
```

```python
x_reshaped.squeeze()
```

```python
x_reshaped.squeeze().shape
```

```python
#torch.unsqueeze() -> adds a single dimension to a target tensor at a specific dim (dimension)

print(f"Previous target : {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

# Add an extra dimension with unsqueeze
x_unsq = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor : {x_unsq}")
print(f"New shape: {x_unsq.shape}")
```

```python
# torch.permute -> rearranges the dimension of a target tensor in a specified order
x_original = torch.rand(size=(224,224,3)) # [height, width, colour_channels]


# Permute the orginal tensor to rearrange the axis (or dim) order
x_per = x_original.permute(2,0,1) # shifts axis 0->2, 1->0, 2-1

print(f"Previous shape : {x_original.shape}")
print(f"New shape : {x_per.shape}") # [colour_channels, height, width ]
```

```python
# Let's index on the middle bracket (dim=1)
x[0][0]
```

```python
# Let's index on the most inner bracket (last dimension)
x[0][0][0]
```

```python
# You can also use ":" to select "all" of a target dimension
x[:,0]
```

```python
# Get all values of 0th and 1st dimension but only index 1 of 2nd dimension
x[:,:,1]
```

```python
# Get all values of the 0 dimension but only the 1 index value of 1srt and 2nd dimension
x[:,1,1]
```

```python
# Get index of 0 of 0th and 1st dimension and all values of 2nd dimension
x[0,0,:]
```

```python
# Index on x to return 9
print(x[0][2][2])

# Index on x to return 3, 6, 9
print(x[:,:,2])
```

# PyTorch tensor & Numpy

Numpy is a popular scientific Python numerical computing library.
And because of this, PyTorch has functionality to interact with it.
* Data in NumPy, want in PyTorch tensor -> `torch.from_numpy(ndarray)`
* PyTorch tensor -> Numpy -> `torch.Tensor.numpy()`

```python
# Numpy array to tensor
import torch
import numpy as np

array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)
array, tensor
```

```python
# Tensor to Numpy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```
